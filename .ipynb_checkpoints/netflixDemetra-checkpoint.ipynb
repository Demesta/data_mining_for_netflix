```python
df10 = df.copy()


df10 = df10[['type', 'duration', 'title']]

df10.query('type == "TV Show"', inplace = True)

df10['duration'] = df10['duration'].str.replace('Seasons', '')
df10['duration'] = df10['duration'].str.replace('Season', '')

df10['duration'] = pd.to_numeric(df10['duration'])



ax = df10.plot.bar(x='title', y='duration', rot=0)


```
